Write OOP classes to handle the following scenarios:
1. A user can create and view 2D coordinates
2. A user can find out the distance between 2 coordinates
3. A user can find find the distance of a coordinate from origin
4. A user can check if a point lies on a given line
5. A user can find the distance between a given 2D point and a given line
6. Also check if two lines are parallel or not (whether they intersect or not)

```python
class Point:
    
    def __init__(self, x, y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self):
        return f"<{self.x_cod}, {self.y_cod}>"

    def euclidean_distance(self, other):
        return ((self.x_cod - other.x_cod) ** 2 + (self.y_cod - other.y_cod) ** 2) ** 0.5

    def distance_from_origin(self):
        return self.euclidean_distance(Point(0, 0))
```

```python
p1 = Point(0, 0)
print(p1)

p2 = Point(10, 10)
print(p2)

print(p1.euclidean_distance(p2))
print(p2.distance_from_origin())
```

```python
class Line:

    def __init__(self, A, B, C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return f"{self.A}X + {self.B}Y + {self.C} = 0"
    
    def point_on_line(self, p: Point) -> bool:
        val = self.A * p.x_cod + self.B * p.y_cod + self.C
        if val == 0:
            return "Lies on the Line"
        else:
            return "Does not lie on the Line"
        
    def shortest_distance(self, p: Point) -> int:
        return abs(self.A * p.x_cod + self.B * p.y_cod + self.C) / ((self.A ** 2 + self.B ** 2) ** 0.5)
    
    def check_parallel(self, l: Line) -> bool:
        # instead check for A1​ * B2​ = A2 * ​B1​
        # slope1 = (-self.A / self.B)
        # slope2 = (-l.A / l.B)
        return self.A * l.B == l.A * self.B
    
    def check_coincident(self, l: "Line") -> bool:
        return (
            self.A * l.B == l.A * self.B and
            self.A * l.C == l.A * self.C and
            self.B * l.C == l.B * self.C
        )
```

```python
l1 = Line(1, 1, -2)
print(l1)
print(l1.point_on_line(Point(1, 1)))
print(l1.point_on_line(Point(0, 1)))
```

```python
print(l1.shortest_distance(Point(1, 10)))
```

```python
l2 = Line(2, 3, 5)
l3 = Line(2, 3, -10)
l4 = Line(2, 8, 7)

print(l2.check_parallel(l3))
print(l2.check_parallel(l4))
```

## How objects access attributes

```python
class Person:
    def __init__(self, name, country):
        self.name = name 
        self.country = country

    def greet(self):
        if self.country == 'India':
            print("Namaste! ", self.name)
        else:
            print("Hello! ", self.name)
```

```python
p = Person('Ram', 'India')
p.greet()
p.name
```

## Reference Variable

1. Reference Variables
2. Reference variables hold the objects
3. We can create objects without reference variable as well
4. An object can have multiple reference variables
5. Assigning a new reference variable to an existing object does not create a new object

```python
# object without a reference

class Person:
    def __init__(self):
        self.name = 'Nitish'
        self.gender = 'Male'

# creating an object without reference
Person()

# multiple reference
p = Person()
q = p        # aliasing
# p is not our object, p and q are just the reference variables which store the reference/ address to the real object created in the memory
```

```python
# multiple reference

print(id(p))
print(id(q))
```

```python
print(p.name)
print(q.name)

q.name = 'Ankit'

print(p.name)
print(q.name)
```
