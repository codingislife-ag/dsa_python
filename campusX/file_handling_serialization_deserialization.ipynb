Some Theory

Types of data used for I/O:

1. Text - '12345' as a sequence of unicode chars
2. Binary - 12345 as a sequence of bytes of its binary equivalent

Hence there are 2 file types to deal with

1. Text files - All program files are text files
2. Binary Files - Images,music,video,exe files

How File I/O is done in most programming languages
1. Open a file
2. Read/Write data
3. Close the file

### Writing to a File

```python
# Case 1 -> If the file is not present

f = open('sample.txt', 'w')
f.write('Hello World')
print(type(f))
f.close()

# since the file has already been close, therefore the below line wont work.
f.write('Hello')
```

```python
# write multiline strings

f = open('sample1.txt', 'w')
f.write('Hello World!')
f.write('\nHow are you?')
f.close()
```

```python
# Case 2 -> The file is already present.

f = open('sample.txt', 'w')
f.write('Salman Khan')
f.close()
```
