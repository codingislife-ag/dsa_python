### Aggregation (Has-A Relationship)

1. Customer has an address
2. Restaurant has a menu

```python
# when creating an object or customer class, we are sending an object of address class.

class Address:
    def __init__(self, city: str, pin: int, state: str) -> None:
        self.__city = city
        self.pin = pin
        self.state = state

    def get_city(self):
        return self.__city
    
    def set_city(self, city):
        self.__city = city

    def edit_address(self, city, pin, state) -> None:
        self.__city = city
        self.pin = pin
        self.state = state

class Customer:
    def __init__(self, name: str, address: Address, gender: str) -> None:
        self.name = name 
        self.address = address
        self.gender = gender

    def print_address(self) -> None:
        print(self.address.get_city(), self.address.pin, self.address.state)

    def edit_profile(self, name: str, city: str, pin: int, state: str) -> None:
        self.name = name
        self.address.edit_address(city, pin, state)

add1 = Address('Gurgaon', '122011', 'Haryana')
cust = Customer('Nitish', add1, 'Male')
cust.print_address()

cust.edit_profile('Ankit', 'Mumbai', 111222, 'Maharashtra')
cust.print_address()
```

```python
# constructor example

class Phone:
    def __init__(self, price: int, brand: str, camera) -> None:
        print('Inside Phone Constructor')
        self.price = price 
        self.brand = brand
        self.camera = camera

    def buy(self) -> None:
        print("Buying a phone")

class SmartPhone(Phone):
    pass

s = SmartPhone(2000, 'Apple', 13)
s.buy()
```

```python
# constructor example 2

class Phone:
    def __init__(self, price, brand, camera) -> None:
        print("Inside Phone Constructor")
        self.price = price
        self.brand = brand
        self.camera = camera
    
    def buy(self) -> None:
        print("Buying a phone")

class SmartPhone(Phone):
    def __init__(self, os, ram) -> None:
        print("Inside SmartPhone Constructor")
        self.os = os
        self.ram = ram
    
    def buy(self) -> None:
        print("Buying a SmartPhone")

s = SmartPhone('Android', 2)
s.buy()
```

```python
# child object cant access the private members of the Parent class

class Phone:
    def __init__(self, price, brand, camera) -> None:
        print("Inside Phone Constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
    
    def __show(self) -> None:
        print(self.__price)

class SmartPhone(Phone):
    def check(self) -> None:
        print(self.__price)

s = SmartPhone(2000, 'Apple', 13)
print(s.brand)
s.check()
```

```python
s.show()
```

#### Use getter to access private members of the Parent Class

```python
class Parent:
    def __init__(self, num):
        self.__num = num
    
    def get_num(self):
        return self.__num

class Child(Parent):

    def show(self) -> None:
        print("This is in child class")

son = Child(100)
print(son.get_num())
son.show()
```

```python
class Parent:
    def __init__(self, num):
        self.__num = num
    
    def get_num(self):
        return self.__num

class Child(Parent):
    def __init__(self, val, num) -> None:
        self.__val = val

    def get_val(self) -> None:
        return self.__val

son = Child(100, 10)
print("Parent: Num: ", son.get_num())
print("Child: Val: ", son.get_val())
```

### Super Keyword

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a Phone")

class SmartPhone(Phone):
    def buy(self) -> None:
        print("Buying a SmartPhone")
        # syntax to call parent's buy method
        super().buy()

s = SmartPhone(20000, "Apple", 13)
s.buy()
s.super().buy()

# super cannot be used outside class
```

```python
# can super access parent ka data?

class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
        # syntax to call parent ka buy method
        print(super().brand)

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

```python
# super -> constructor

class Phone:
    def __init__(self, price: int, brand: str, camera) -> None:
        print("Inside Phone Constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, price: int, brand: str, camera, os: str, ram: int) -> None:
        print("Inside SmartPhone Constructor")
        super().__init__(price, brand, camera)
        self.os = os 
        self.ram = ram
        print("Inside SmartPhone Constructor")

s = SmartPhone(20000, 'Apple', 13, 'IOS', 2)
print(s.os)
print(s.brand)
```

Inheritance in summary
A class can inherit from another class.

Inheritance improves code reuse

Constructor, attributes, methods get inherited to the child class

The parent has no access to the child class

Private properties of parent are not accessible directly in child class

Child class can override the attributes or methods. This is called method overriding

super() is an inbuilt function which is used to invoke the parent class methods and constructor
